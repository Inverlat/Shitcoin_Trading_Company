```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("../Binance_Data_Extractor/BTCUSDT.csv")
```

```python
df = df[["Open_Time", "Open", "High", "Low", "Close", "Volume", "Number_Of_Trades"]]
# Nos indica la tendencia
df["Direction"] = (df["Close"] >= df["Open"]).astype(int)*2-1
df["SMA_100"] = df["Close"].rolling(100).mean()
df["SMA_100"] = df["SMA_100"].fillna(0)
df["SMA_30"] = df["Close"].rolling(30).mean()
df["SMA_30"] = df["SMA_30"].fillna(0)
# Si la SMA a corto plazo es mayor a la de largo entonces el mercado va largo
df["Long"] = (df["SMA_30"] > df["SMA_100"]).astype(int)
df
# Aber si ves como arreglar estas putas advertencias
```
